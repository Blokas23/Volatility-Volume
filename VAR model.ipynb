## Stationarity Test

According to the documentation of statsmodels we can only use the VAR class if we have tested for stationarity.

```python
def adfuller_test(series, signif=0.05, name='', verbose=False):
    """Perform ADFuller to test for Stationarity of given series and print report"""
    r = adfuller(series, autolag='AIC')
    output = {'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
    p_value = output['pvalue'] 
    def adjust(val, length= 6): return str(val).ljust(length)

    # Print Summary
    print(f'    Augmented Dickey-Fuller Test on "{name}"', "\n   ", '-'*47)
    print(f' Null Hypothesis: Data has unit root. Non-Stationary.')
    print(f' Significance Level    = {signif}')
    print(f' Test Statistic        = {output["test_statistic"]}')
    print(f' No. Lags Chosen       = {output["n_lags"]}')

    for key,val in r[4].items():
        print(f' Critical value {adjust(key)} = {round(val, 3)}')

    if p_value <= signif:
        print(f" => P-Value = {p_value}. Rejecting Null Hypothesis.")
        print(f" => Series is Stationary.")
    else:
        print(f" => P-Value = {p_value}. Weak evidence to reject the Null Hypothesis.")
        print(f" => Series is Non-Stationary.")
```

## Granger Causality Test (predictability)

Granger Causality will show us what to expect from each of the variables in terms of predicting other variables in the model.

```python
maxlag=12
test = 'ssr_chi2test'
def grangers_causation_matrix(data, variables, test='ssr_chi2test', verbose=False):    
    """Check Granger Causality of all possible combinations of the Time series.
    The rows are the response variable, columns are predictors. The values in the table 
    are the P-Values. P-Values lesser than the significance level (0.05), implies 
    the Null Hypothesis that the coefficients of the corresponding past values is 
    zero, that is, the X does not cause Y can be rejected.

    data      : pandas dataframe containing the time series variables
    variables : list containing names of the time series variables.
    """
    df = pd.DataFrame(np.zeros((len(variables), len(variables))), columns=variables, index=variables)
    for c in df.columns:
        for r in df.index:
            test_result = grangercausalitytests(data[[r, c]], maxlag=maxlag, verbose=False)
            p_values = [round(test_result[i+1][0][test][1],4) for i in range(maxlag)]
            if verbose: print(f'Y = {r}, X = {c}, P Values = {p_values}')
            min_p_value = np.min(p_values)
            df.loc[r, c] = min_p_value
    df.columns = [var + '_x' for var in variables]
    df.index = [var + '_y' for var in variables]
    return df

grangers_causation_matrix(VAR_data, variables = VAR_data.columns)
```

## Cointegration Test

Cointegration test helps to establish the presence of a statistically significant connection between two or more time series.

But, what does Cointegration mean?

To understand that, you first need to know what is ‘order of integration’ (d).

Order of integration(d) is nothing but the number of differencing required to make a non-stationary time series stationary.

Now, when you have two or more time series, and there exists a linear combination of them that has an order of integration (d) less than that of the individual series, then the collection of series is said to be cointegrated.

We can implement Johansen test

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen

def cointegration_test(df, alpha=0.05): 
    """Perform Johanson's Cointegration Test and Report Summary"""
    out = coint_johansen(df,-1,5)
    d = {'0.90':0, '0.95':1, '0.99':2}
    traces = out.lr1
    cvts = out.cvt[:, d[str(1-alpha)]]
    def adjust(val, length= 6): return str(val).ljust(length)

    # Summary
    print('Name   ::  Test Stat > C(95%)    =>   Signif  \n', '--'*20)
    for col, trace, cvt in zip(df.columns, traces, cvts):
        print(adjust(col), ':: ', adjust(round(trace,2), 9), ">", adjust(cvt, 8), ' =>  ' , trace > cvt)

cointegration_test(VAR_data)
```

## Structural Breaks

 The validation and train set seem to be wildly different to each other. We will perform a test to confirm these suspicions in the form of a Chow multivariate test.

The test statistic is of the form

$$ \text{CHOW} = {\frac{\left( 𝑆𝑄𝐸_𝑟− (𝑆𝑄𝐸_1 + 𝑆𝑄𝐸_2) \right) / 𝑘}{(𝑆𝑄𝐸_1+𝑆𝑄𝐸_2) / (𝑁−2𝑘)}}$$	

Here 
1. $SQE_r$ is the sum of squared residuals of the main model (pre + post)
2. $SQE_1$ is the sum of squared residuals of the train model
3. $SQE_2$ is the sum of squared residuals of the valid model
4. $N$ is total number of observations train + valid
5. $k$ is number of predictors (including the constant of the lin. reg.)

We then compare the CHOW statistic with the corresponding $F(N,k)$ critical value since the CHOW statistic behaves according to the F distribution.

```python
# For the implied _ realized dataset
X_pre_covid = sm.add_constant(((VAR_data[0:1015].drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_pre_covid = ((VAR_data[0:1015].drop(columns=['implied_volatility','realized_volatility'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
X_post_covid = sm.add_constant(((VAR_data[1015:].drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_post_covid  = ((VAR_data[1015:].drop(columns=['implied_volatility','realized_volatility'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
X_total = sm.add_constant(((VAR_data.drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_total = ((VAR_data.drop(columns=['implied_volatility','realized_volatility'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
```

```python
# For the difference dataset
X_pre_covid_diff = sm.add_constant(((VAR_data_diff[0:1015].drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_pre_covid_diff = ((VAR_data_diff[0:1015].drop(columns=['difference'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
X_post_covid_diff = sm.add_constant(((VAR_data_diff[1015:].drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_post_covid_diff  = ((VAR_data_diff[1015:].drop(columns=['difference'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
X_total_diff = sm.add_constant(((VAR_data_diff.drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_total_diff = ((VAR_data_diff.drop(columns=['difference'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
```

```python
# For the Benchmark dataset
# For the implied _ realized dataset
X_pre_covid_benchmark = sm.add_constant(((VAR_data_benchmark[0:1015].drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_pre_covid_benchmark = ((VAR_data_benchmark[0:1015].drop(columns=['implied_volatility','realized_volatility'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
X_post_covid_benchmark = sm.add_constant(((VAR_data_benchmark[1015:].drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_post_covid_benchmark  = ((VAR_data_benchmark[1015:].drop(columns=['implied_volatility','realized_volatility'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
X_total_benchmark = sm.add_constant(((VAR_data_benchmark.drop(columns='Volume',inplace=False)).reset_index()).drop(columns="Date",inplace=False))
Y_total_benchmark = ((VAR_data_benchmark.drop(columns=['implied_volatility','realized_volatility'],inplace=False)).reset_index()).drop(columns="Date",inplace=False)
```

### Chow test calculated data

```python
model_pre = sm.OLS(Y_pre_covid,X_pre_covid)
model_post = sm.OLS(Y_post_covid,X_post_covid)
model_full = sm.OLS(Y_total,X_total)
result_pre = model_pre.fit()
result_post = model_post.fit()
result_full = model_full.fit()

SQER = ((result_full.resid)**2).sum()
SQE1 = ((result_pre.resid)**2).sum()
SQE2 = ((result_post.resid)**2).sum()
k = 2
N = 1141

Numerator = (SQER - (SQE1 + SQE2))/k
Denominator = (SQE1 + SQE2) / (N-2*k)

Chow_statistic = Numerator / Denominator
F_table_value = scipy.stats.f.ppf(q=1-0.05, dfn=2, dfd=1141)

print("CHOW TEST: ")
if Chow_statistic>F_table_value:
    print("Chow statistic is "+ str(round(Chow_statistic,2)))
    print("Table value of the F distribution is "+ str(round(F_table_value,2))) 
    print("We can confirm the structural break on 95% confidence.")
else:
    print("We cannot confirm the structural break on 95% confidence.")
```

### Chow test difference data

```python
model_pre_diff = sm.OLS(Y_pre_covid_diff,X_pre_covid_diff)
model_post_diff = sm.OLS(Y_post_covid_diff,X_post_covid_diff)
model_full_diff = sm.OLS(Y_total_diff,X_total_diff)
result_pre_diff = model_pre_diff.fit()
result_post_diff = model_post_diff.fit()
result_full_diff = model_full_diff.fit()

SQER_diff = ((result_full_diff.resid)**2).sum()
SQE1_diff = ((result_pre_diff.resid)**2).sum()
SQE2_diff = ((result_post_diff.resid)**2).sum()
k = 2
N = 1141

Numerator = (SQER_diff - (SQE1_diff + SQE2_diff))/k
Denominator = (SQE1_diff + SQE2_diff) / (N-2*k)

Chow_statistic = Numerator / Denominator
F_table_value = scipy.stats.f.ppf(q=1-0.05, dfn=2, dfd=1141)

print("CHOW TEST: ")
if Chow_statistic>F_table_value:
    print("Chow statistic is "+ str(round(Chow_statistic,2)))
    print("Table value of the F distribution is "+ str(round(F_table_value,2))) 
    print("We can confirm the structural break on 95% confidence.")
else:
    print("We cannot confirm the structural break on 95% confidence.")
```

```python
def coefplot(results):
    '''
    Takes in results of OLS model and returns a plot of 
    the coefficients with 95% confidence intervals.
    
    Removes intercept, so if uncentered will return error.
    '''
    # Create dataframe of results summary 
    coef_df = pd.DataFrame(results.summary().tables[1].data)
    
    # Add column names
    coef_df.columns = coef_df.iloc[0]

    # Drop the extra row with column labels
    coef_df=coef_df.drop(0)

    # Set index to variable names 
    coef_df = coef_df.set_index(coef_df.columns[0])

    # Change datatype from object to float
    coef_df = coef_df.astype(float)

    # Get errors; (coef - lower bound of conf interval)
    errors = coef_df['coef'] - coef_df['[0.025']
    
    # Append errors column to dataframe
    coef_df['errors'] = errors

    # Drop the constant for plotting
    coef_df = coef_df.drop(['const'])

    # Sort values by coef ascending
    coef_df = coef_df.sort_values(by=['coef'])

    ### Plot Coefficients ###

    # x-labels
    variables = list(coef_df.index.values)
    
    # Add variables column to dataframe
    coef_df['variables'] = variables
    
    # Set sns plot style back to 'poster'
    # This will make bars wide on plot
    sns.set_context("poster")

    # Define figure, axes, and plot
    fig, ax = plt.subplots(figsize=(6, 4))
    
    # Error bars for 95% confidence interval
    # Can increase capsize to add whiskers
    coef_df.plot(x='variables', y='coef', kind='bar',
                 ax=ax, color='none', fontsize=14, 
                 ecolor='steelblue',capsize=0,
                 yerr='errors', legend=False)
    
    # Set title & labels
    plt.title('Coefficients of Features w/ 95% Confidence Intervals',fontsize=16)
    ax.set_ylabel('Coefficients',fontsize=14)
    ax.set_xlabel('',fontsize=14)
    for tick in ax.get_xticklabels():
        tick.set_rotation(25)
    
    # Coefficients
    ax.scatter(x=pd.np.arange(coef_df.shape[0]), 
               marker='o', s=80, 
               y=coef_df['coef'], color='steelblue')
    
    # Line to define zero on the y-axis
    ax.axhline(y=0, linestyle='--', color='red', linewidth=1)
    plt.savefig('precovidcoeff.png', bbox_inches='tight')
    return plt.show()
```
