Setting up Christmas Dummy

```python
volume['Christmas'] = 0
listOfStrings = ['12/25','12/26','12/27','12/28','12/29','12/30','12/31','01/01','01/02']
for i in range(0,len(volume)):
    if volume.index[i][0:5] in listOfStrings:
        volume.at[volume.index[i], 'Christmas'] = 1
```

Setting up Mondays Dummy

```python
volume['Date'] = pd.to_datetime(volume.index)
volume['weekdays'] = volume['Date'].dt.day_name()
volume['is_Monday'] = np.where(volume['weekdays'] == 'Monday', 1,0)
```

Setting up the t- variables

```python
volume['Counter'] = 0
for i in range(0,len(volume)):
    volume.at[volume.index[i], 'Counter'] = i
```

```python
def minmax_norm(df_input):
    return (df_input - df_input.min()) / ( df_input.max() - df_input.min())
```

```python
volume['t'] = minmax_norm(volume['Counter'])
volume['t^2'] = volume['t']**2
```

```python
Concat = np.concatenate((a,b,c,d,e,f,g,i,j,k,l,m,n,o,p,r,s,t,u,v))
volume["TTMt"] = Concat
volume["TTMt^2"] =volume["TTMt"] **2
```

Generating the $cos$ and $sin$ elements

```python
lookback = list(range(1,6))
```

```python
for i in lookback:
    volume[f'CosineElement{i}'] = np.cos(2*np.pi*i*volume['TTMt']/TTMmax)
```

```python
for i in lookback:
    volume[f'Sinelement{i}'] = np.sin(2*np.pi*i*volume['TTMt']/TTMmax)
```

```python
Volume = volume.TC
#VolumeOI = volume.VolumeOI
Features = volume [['Christmas','is_Monday','t','t^2','TTMt','TTMt^2','CosineElement1','CosineElement2','CosineElement3','CosineElement4','CosineElement5',
                   'Sinelement1','Sinelement2','Sinelement3','Sinelement4','Sinelement5']]
Features = sm.add_constant(Features,prepend=True)
```

```python
#est=sm.OLS(VolumeOI, Features)
est = sm.OLS(Volume,Features)
est = est.fit()
est.summary()
```

## Aikake Information Criterium

The formula is $AIC = n * LL + 2 * k$ <br />

LL is the log likelihood, in practice natural log of RMSE , k is the number of parameters.
https://machinelearningmastery.com/probabilistic-model-selection-measures/

```python
# calculate aic for regression
def calculate_aic(n, mse, num_params):
    aic = n * log(mse) + 2 * num_params
    return aic
```

```python
aic = []
from sklearn.metrics import mean_squared_error
from math import log
```

```python
for j in range(1,30):
    Features_AIC = volume [['Christmas','is_Monday','t','t^2','TTMt','TTMt^2']]
    for i in range(1,j):
        Features_AIC[f'CosineElement{i}'] = np.cos(2*np.pi*i*Features_AIC['TTMt']/TTMmax)
        Features_AIC[f'Sinelement{i}'] = np.sin(2*np.pi*i*Features_AIC['TTMt']/TTMmax)
    Features_AIC = sm.add_constant(Features_AIC,prepend=True)   
    est=sm.OLS(Volume, Features_AIC)
    est = est.fit()
    yhat = est.predict(Features_AIC)
    num_params = len(est.params) + 1
    mse = mean_squared_error(Volume, yhat)
    aic_ = calculate_aic(len(Volume), mse, num_params)
    aic.append(aic_)
```

```python
fig = plt.figure()
ax = plt.axes()

x = range(1,30)
ax.plot(x, aic)
```
